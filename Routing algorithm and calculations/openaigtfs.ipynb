```python
import pandas as pd
import zipfile

# Path to your GTFS ZIP file
gtfs_zip_path = 'agency.zip'

# Name of the file within the ZIP archive
file_name = 'shapes.txt'

# Read shapes.txt from the ZIP archive into a pandas DataFrame
with zipfile.ZipFile(gtfs_zip_path, 'r') as zip_ref:
    with zip_ref.open(file_name) as file:
        shapes_df = pd.read_csv(file)

# Display the first few rows of the DataFrame
print(shapes_df.head())
```
